```python
df['text'].fillna('', inplace=True)
```

Using Spacy

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Define a function to perform POS tagging
def pos_tagging(text):
    doc = nlp(text)
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags

# Apply the function to the 'text' column
df['pos_tags'] = df['text'].apply(pos_tagging)
```

```python
df.head()
```

```python
rows_with_label_0 = df[df['label'] == 0]

print(rows_with_label_0)
```

```python
# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Using the spaCy dependency parser
# doc = nlp('technically that is still turning back the clock dick head')
doc = nlp('american of all age and race agree trump is a racist resist resistrump fucktrump')
dependency_features = ['Id', 'Text', 'Head', 'Dep']
head_format = "\033[1m{!s:>11}\033[0m" * (len(dependency_features) )
row_format = "{!s:>11}" * (len(dependency_features) )

print(head_format.format(*dependency_features))
# Printing dependency features for each token
for token in doc:
    print(row_format.format(token.i, token.text, token.head.i, token.dep_))
```

```python
# SpaCy visualization tool
from spacy import displacy

# Run in a terminal
displacy.serve(doc, style='dep')
```

Constituency Parsing

```python
import stanza

nlp = stanza.Pipeline(lang='en', processors='tokenize,pos,constituency')
doc = nlp('american of all age and race agree trump is a racist resist resistrump fucktrump')
```

```python
for sentence in doc.sentences:
    print(sentence.constituency)
```
